### Activity 11: Create a new dictionary of teams from players_data

Create a dictionary teams with the unique teams in the players_data dictionary as the key and the number of players in each team as the value.

<code>
{'Los Angeles Lakers': 2,
 'Los Angeles Clippers': 2,
 'Milwaukee Bucks': 3,
 'Brooklyn Nets': 2,
...
}
</code>

```python
teams = {}
for key, values in players_data.items():
    if values['team'] not in teams:
        teams[values['team']] = 1
    else:
        current_team_count = teams[values['team']]
        teams[values['team']] = current_team_count + 1

teams
```
